```python
import spacy
nlp = spacy.load("en_core_web_sm")

with open('./data/alice.txt') as f:
    txt = f.read()
```

# 30. POS Tagging

```python
sentences = txt.replace('\n\n', '.').split('.')
sentences = list(filter(lambda s: s != '', map(lambda s: s.strip().replace('\n', ' '), sentences)))
```

```python
docs = [nlp(s) for s in sentences]
```

```python
for token in docs[0]:
    print(token.text, token.lemma_, token.pos_)
```

# 31. Extract Verbs

```python
verb_tokens = []
for doc in docs:
    for token in doc:
        if token.pos_ == "VERB":
            print(token.text, token.lemma_, token.pos_)
            verb_tokens.append(token)
```

# 32. Lemmas

```python
for token in verb_tokens:
    print(token.lemma_)
```

# 33. A of B

```python
# https://stackoverflow.com/questions/6822725/rolling-or-sliding-window-iterator
import more_itertools

for doc in docs:
    triplets = list(more_itertools.windowed(doc, n=3))
    for t in triplets:
        if all(t) and t[0].pos_ == 'NOUN' and t[1].text == "of" and t[2].pos_ == 'NOUN':
            print(t, t[0].pos_, t[1].pos_, t[2].pos_)
```

# 35. Frequency of Words

* Flatten a list of lists: https://stackoverflow.com/questions/952914/how-to-make-a-flat-list-out-of-list-of-lists
* Remove non characters: https://stackoverflow.com/questions/5843518/remove-all-special-characters-punctuation-and-spaces-from-string

```python
from collections import Counter

words = [str(token).lower() for doc in docs for token in doc]
words = list(filter(lambda word: word.isalnum(), words))
c = Counter(words)

l = [(key, c[key]) for key in c]
sorted_list = sorted(l, key=lambda pair: -pair[1])
print(sorted_list[:10])
```

# 37. Co-occuring words

According to [this WikiPedia page](https://en.wikipedia.org/wiki/Co-occurrence), "co-occurence" means:

> In linguistics, co-occurrence or cooccurrence is an above-chance frequency of occurrence of two terms (also known as coincidence or concurrence) from a text corpus alongside each other in a certain order.

```python
import more_itertools

cooccuring = []
for doc in docs:
    pairs = list(more_itertools.windowed(doc, n=2))
    for pair in pairs:
        pair = (str(pair[0]).lower(), str(pair[1]).lower())
        if pair[0] == "alice":
            cooccuring.append(pair[1])
        elif pair[1] == "alice":
            cooccuring.append(pair[0])
```

```python
cooccuring = list(filter(lambda word: word.isalnum(), cooccuring))

c = Counter(cooccuring)
l = [(key, c[key]) for key in c]
sorted_list = sorted(l, key=lambda pair: -pair[1])
print(sorted_list[:10])
```

```python
import seaborn as sns
import pandas as pd
sns.set_theme(style="whitegrid")

df = pd.DataFrame(sorted_list, columns=["word", "count"])
ax = sns.barplot(x="count", y="word", data=df[:10])
```
